```python
import gradio as gr
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
persist_directory = "/content/drive/MyDrive/Colab LLM Project/chroma_db"

openai_embed = OpenAIEmbeddings(api_key="sk-proj-eIwtDvIacKR14ovPvVxs9rJevV-3lGyGt2W5qM4SxfkRV27sgCNoVTiTn1UseuDQjikh9it5ifT3BlbkFJrRjq3hmZ9EK4mcQdsV4x0uCGVffEVXQIeIrGXgrwEUxLZOMN-XXXXXXXXX")

vector_db = Chroma(
    persist_directory=persist_directory,
    embedding_function=openai_embed
)

llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    temperature=0,
    api_key="sk-proj-eIwtDvIacKR14ovPvVxs9rJevV-3lGyGt2W5qM4SxfkRV27sgCNoVTiTn1UseuDQjikh9it5ifT3BlbkFJrRjq3hmZ9EK4mcQdsV4x0uCGVffEVXQIeIrGXgrwEUxLZOMN-XXXXXXXXXXXXXXX"
)

retriever = vector_db.as_retriever(search_kwargs={"k": 3})

prompt = ChatPromptTemplate.from_template(
"""
Answer the question using the context below.

Context:
{context}

Question:
{input}
"""
)
```

```python
def fnQuery(query):
  docs = retriever.invoke(query)
  context = "\n\n".join(doc.page_content for doc in docs)
  formatted_prompt = prompt.format(
                    context=context,
                    input=query
                    )
  response = llm.invoke(formatted_prompt)
  return response.content

demo = gr.Interface(fn=fnQuery, inputs=gr.Textbox(label="Question"), outputs=gr.Markdown(), flagging_mode="never")
demo.launch()
```
